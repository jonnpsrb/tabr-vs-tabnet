```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Given AUC and Accuracy scores for 5 models across 5 datasets
auc_scores = [
    [0.865, 0.883, 0.883, 0.880, 0.860],  # Dataset 1 for each model
    [0.770, 0.853, 0.848, 0.850, 0.789],  # Dataset 2
    [0.913, 0.914, 0.958, 0.983, 0.881],  # Dataset 3
    [0.860, 0.872, 0.876, 0.875, 0.862],  # Dataset 4
    [0.927, 0.928, 0.922, 0.934, 0.917],  # Dataset 5
]

accuracy_scores = [
    [0.79, 0.80, 0.81, 0.81, 0.78],  # Dataset 1 for each model
    [0.77, 0.77, 0.77, 0.77, 0.72],  # Dataset 2
    [0.84, 0.84, 0.89, 0.93, 0.80],  # Dataset 3
    [0.78, 0.80, 0.80, 0.80, 0.79],  # Dataset 4
    [0.85, 0.85, 0.85, 0.87, 0.84],  # Dataset 5
]

# Convert data to DataFrames for better labeling in heatmaps
df_auc = pd.DataFrame(
    auc_scores,
    columns=['MLP', 'MLP-PLR', 'XGBoost', 'TabR', 'TabNet'],
    index=['bank-marketing', 'credit', 'electricity', 'jannis', 'MagicTelescope']
)

df_accuracy = pd.DataFrame(
    accuracy_scores,
    columns=['MLP', 'MLP-PLR', 'XGBoost', 'TabR', 'TabNet'],
    index=['bank-marketing', 'credit', 'electricity', 'jannis', 'MagicTelescope']
)

# Plotting two separate heatmaps: one for AUC and one for Accuracy
plt.figure(figsize=(12, 5))

# AUC heatmap
plt.subplot(1, 2, 1)
sns.heatmap(df_auc, annot=True, cmap="YlGnBu", cbar_kws={'label': 'AUC Score'})
plt.title("AUC Scores for Models Across Datasets")
plt.xlabel("Models")
plt.ylabel("Datasets")

# Accuracy heatmap
plt.subplot(1, 2, 2)
sns.heatmap(df_accuracy, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Accuracy Score'})
plt.title("Accuracy Scores for Models Across Datasets")
plt.xlabel("Models")
plt.ylabel("Datasets")

plt.tight_layout()
plt.show()

```
