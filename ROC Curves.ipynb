```python
import pandas as pd
import matplotlib.pyplot as plt
import os

files = ['bank-marketing-MLP.csv', 'bank-marketing-MLP-PLR.csv', 'bank-marketing-XGBoost.csv', 'bank-marketing-TabR.csv', 'bank-marketing-TabNet.csv',
         'credit-MLP.csv', 'credit-MLP-PLR.csv', 'credit-XGBoost.csv', 'credit-TabR.csv', 'credit-TabNet.csv',
         'electricity-MLP.csv', 'electricity-MLP-PLR.csv', 'electricity-XGBoost.csv', 'electricity-TabR.csv', 'electricity-TabNet.csv'
         ]

auc_scores = {
    'bank-marketing-MLP.csv': 0.86,
    'bank-marketing-MLP-PLR.csv': 0.88,
    'bank-marketing-XGBoost.csv': 0.88,
    'bank-marketing-TabR.csv': 0.88,
    'bank-marketing-TabNet.csv': 0.86,
    'credit-MLP.csv': 0.77,
    'credit-MLP-PLR.csv': 0.85,
    'credit-XGBoost.csv': 0.85,
    'credit-TabR.csv': 0.85,
    'credit-TabNet.csv': 0.79,
    'electricity-MLP.csv': 0.91,
    'electricity-MLP-PLR.csv': 0.91,
    'electricity-XGBoost.csv': 0.96,
    'electricity-TabR.csv': 0.98,
    'electricity-TabNet.csv': 0.88,
}

datasets = {
    'bank-marketing': ['bank-marketing-MLP.csv', 'bank-marketing-MLP-PLR.csv', 'bank-marketing-XGBoost.csv', 'bank-marketing-TabR.csv', 'bank-marketing-TabNet.csv'],
    'credit': ['credit-MLP.csv', 'credit-MLP-PLR.csv', 'credit-XGBoost.csv', 'credit-TabR.csv', 'credit-TabNet.csv'],
    'electricity': ['electricity-MLP.csv', 'electricity-MLP-PLR.csv', 'electricity-XGBoost.csv', 'electricity-TabR.csv', 'electricity-TabNet.csv']
}

fig, axes = plt.subplots(1, len(datasets), figsize=(15, 6), sharey=True)

for ax, (dataset, dataset_files) in zip(axes, datasets.items()):
    for file in dataset_files:
        df = pd.read_csv(file)
        fpr = df['fpr']
        tpr = df['tpr']

        auc_score = auc_scores.get(file, None)

        label = os.path.basename(file).replace('.csv', '')
        if auc_score is not None:
            label += f' (AUC = {auc_score:.2f})'

        ax.plot(fpr, tpr, label=label)

    ax.plot([0, 1], [0, 1], color='gray', linestyle='--', label='Random classifier')

    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_title(f'ROC Curve - {dataset}')
    ax.set_xlabel('False Positive Rate')
    ax.legend(loc='lower right')

fig.text(0, 0.5, 'True Positive Rate', va='center', rotation='vertical')

plt.tight_layout()
plt.show()
```
